```
import pandas as pd
from urllib.request import urlretrieve
url = "https://github.com/Elwing-Chou/ml0804/raw/master/house_train.csv"
urlretrieve(url, "house_train.csv")
train_df = pd.read_csv("house_train.csv", encoding="utf-8")
url = "https://github.com/Elwing-Chou/ml0804/raw/master/house_test.csv"
urlretrieve(url, "house_test.csv")
```

```
import pandas as pd
train_df = pd.read_csv("house_train.csv", encoding="utf-8")
train_df
```

```
test_df = pd.read_csv("house_test.csv", encoding="utf-8")
test_df
```

```
# concat: axis!!!
datas = pd.concat([train_df, test_df]).reset_index(drop=True)
datas = datas.drop(["SalePrice"], axis=1)
datas
```

```
s = datas.isna().sum()
s[s > 0].sort_values(ascending=True)
```

```
def Alley(s):
  if not pd.isna(s):
    return s[0]
  else:
    return None
datas["Alley"] = datas["Alley"].apply(Alley))
```
